## 미세먼지 null 값 대체하기

```python
%cd /content/drive/MyDrive/KOPIS DATA/Data/csv/Outsource/Weather/microdust/
import pandas as pd
import math

df = pd.read_csv('micro_dust_2019_2022.csv',encoding='cp949')

# '구분'이 '평균'인 행 제거하기
df = df[df['구분'] != '평균']
df.head()
```

```python
# 날짜와 구분을 기준으로 데이터프레임을 정렬
df = df.sort_values(by=['구분', '일시']).reset_index(drop=True)

# null 값을 가장 최근의 유효한 값으로 대체
for idx, row in df.iterrows():
    if pd.isnull(row['미세먼지(PM10)']):
        prev_valid = df.loc[:idx, '미세먼지(PM10)'].dropna().iloc[-1]
        next_valid = df.loc[idx:, '미세먼지(PM10)'].dropna().iloc[0]
        df.at[idx, '미세먼지(PM10)'] = (prev_valid + next_valid) / 2

    if pd.isnull(row['초미세먼지(PM2.5)']):
        prev_valid = df.loc[:idx, '초미세먼지(PM2.5)'].dropna().iloc[-1]
        next_valid = df.loc[idx:, '초미세먼지(PM2.5)'].dropna().iloc[0]
        df.at[idx, '초미세먼지(PM2.5)'] = (prev_valid + next_valid) / 2

print(df)
```

## 미세먼지등급 분류하기

```python
# 미세먼지와 초미세먼지 등급을 반환하는 함수 정의
def get_pm10_grade(value):
    if math.isnan(value):
        return None
    elif value < 30:
        return 0
    elif value < 80:
        return 1
    elif value < 150:
        return 2
    else:
        return 3

def get_pm25_grade(value):
    if math.isnan(value):
        return None
    elif value < 15:
        return 0
    elif value < 35:
        return 1
    elif value < 75:
        return 2
    else:
        return 3

# '미세먼지등급'과 '초미세먼지등급' 칼럼 추가
df['미세먼지등급'] = df['미세먼지(PM10)'].apply(get_pm10_grade)
df['초미세먼지등급'] = df['초미세먼지(PM2.5)'].apply(get_pm25_grade)

df.to_csv('/content/drive/MyDrive/KOPIS DATA/Data/csv/Outsource/Weather/microdust/micro_dust_level.csv', index=False)
```

```python
def extract_first_text_inside_parentheses(input_string):
    pattern = r'\(([^)]+)'  # 정규 표현식: 첫 번째 소괄호 안의 모든 문자를 추출
    result = re.search(pattern, str(input_string))
    if result: #소괄호가 여러번 나오는 주소가 있어 가장 먼저 나오는 소괄호만 유효하도록 함
        return result.group(1)
    return None

seoul['행정동'] = seoul['주소'].apply(extract_first_text_inside_parentheses)
```

```python
import os
import pandas as pd

%cd /content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/Original/서대문구4_날씨_2019_2022


def merge_csvs_by_group(file_list, output_folder):
    # 파일들을 두 번째 언더바까지의 정보를 기준으로 그룹화
    groups = {}
    for csv_file in file_list:
        key = '_'.join(csv_file.split('_')[:2])
        if key not in groups:
            groups[key] = []
        groups[key].append(csv_file)

    # 각 그룹별로 파일들을 병합하여 저장
    for group_key, group_files in groups.items():
        # 빈 DataFrame 생성
        merged_df = pd.DataFrame()
        # 각 파일을 순환하면서 병합
        for csv_file in group_files:
            # 첫 번째 행을 제외하고 읽어옴
            df = pd.read_csv(csv_file)
            df.columns = ["date", "time", "value"]

            # "date" 열에서 "Start"를 포함하는 행을 삭제
            df = df[~df['date'].str.contains('Start')]

            merged_df = pd.concat([merged_df, df], ignore_index=True)

        # 파일 이름에 '2019_2022'를 추가하여 저장
        output_filename = f"{output_folder}/{group_key}_2019_2022.csv"
        merged_df.to_csv(output_filename, index=False)
        print(f"{group_key} 병합 및 저장 완료: {output_filename}")
```

```python
import os
import pandas as pd
import datetime

# Function to convert hour value to 'hh:mm' format
def convert_hour_format(time):
    if pd.notnull(time):
        try:
            hours = int(time) // 100
            minutes = int(time) % 100
            return f"{hours:02d}:{minutes:02d}"
        except ValueError:
            return 'NA'  # Return 'NA' if hour value is not a valid integer
    return 'NA'  # Return 'NA' for NaN values

# Function to perform the column renaming based on filename
def rename_columns(df, filename):
    df.rename(columns={df.columns[0]: 'date', 'time': 'time_temp'}, inplace=True)
    df['time'] = df['time_temp'].apply(convert_hour_format)
    df.drop(columns='time_temp', inplace=True)
    return df

# Function to modify the 'date' column with dates from 2019-07-01 to 2022-12-31
def modify_date_column(df):
    start_date = datetime.date(2019, 7, 1)
    end_date = datetime.date(2022, 12, 31)
    dates = [start_date + datetime.timedelta(days=i) for i in range((end_date - start_date).days + 1)]
    df['date'] = sorted(dates * 24)  # Sort the dates and repeat 24 times for each date
    return df

# Directory where the files are located
directory = '/content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/Merged/merged_서대문구4_날씨'
out_directory = '/content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/Format/format_서대문구_추가'

# Get the list of files in the directory
file_list = os.listdir(directory)

# Iterate through each file
for file in file_list:
    if file.endswith('.csv'):
        # Read the file into a DataFrame
        df = pd.read_csv(os.path.join(directory, file))

        # Get the filename without extension
        filename_without_extension = os.path.splitext(file)[0]

        # Perform column renaming based on filename
        df = rename_columns(df, filename_without_extension)

        # Reorder columns to have 'hour' as the second column
        if 'time' in df.columns:
            cols = list(df.columns)
            cols.insert(1, cols.pop(cols.index('time')))
            df = df[cols]

        # Modify the 'date' column with dates from 2019-07-01 to 2022-12-31
        df = modify_date_column(df)

        # Save the updated DataFrame back to the file
        df.to_csv(os.path.join(out_directory, file), index=False)
```

## 강수 기온 습도 합치기

```python
import os
import pandas as pd

# 데이터가 있는 경로를 지정합니다.
data_path = '/content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/Format/format_서대문구_추가'  # 데이터가 있는 경로로 변경해주세요.

# 데이터를 저장할 결과 디렉토리를 생성합니다.
result_path = '/content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/Final/final_서대문구_날씨_2019_2022'  # 결과를 저장할 경로로 변경해주세요.
os.makedirs(result_path, exist_ok=True)

# 첫 번째 언더바 전까지의 문자가 일치하는 CSV 파일들을 그룹화합니다.
grouped_files = {}
files = os.listdir(data_path)
files.sort()  # Sort filenames
for csv_file in files:
    file_name = csv_file.split('.')[0]
    group_name = '_'.join(file_name.split('_')[:1])
    if group_name not in grouped_files:
        grouped_files[group_name] = []
    grouped_files[group_name].append(csv_file)

# 그룹화된 파일들을 읽어들여서 결과 데이터프레임에 세 번째 열을 추가합니다.
result_df = pd.DataFrame()
for group_name, files in grouped_files.items():
    group_dfs = []
    first_file_read = False
    for idx, csv_file in enumerate(files):
        csv_path = os.path.join(data_path, csv_file)
        df = pd.read_csv(csv_path)
        column_name = csv_file.split('_')[1]  # 첫 번째 언더바와 두 번째 언더바 사이의 문자를 가져옵니다.
        if not first_file_read:
            # 첫 번째 파일의 경우 세 번째 열 이름을 변경합니다.
            df.columns.values[2] = column_name
            group_dfs.append(df)
            first_file_read = True
        else:
            group_dfs.append(df.iloc[:, 2].rename(column_name))
    merged_df = pd.concat(group_dfs, axis=1)
    result_df = pd.concat([result_df, merged_df], axis=0)

    # 새로운 파일 이름을 지정하여 결과를 저장합니다.
    new_file_name = f"{group_name}_강수기온습도_2019_2022.csv"
    result_file_path = os.path.join(result_path, new_file_name)
    merged_df.to_csv(result_file_path, index=False)

print("작업이 완료되었습니다. 결과는", result_path, "에 저장되었습니다.")
```

```python
grouped = sale.groupby('공연일시')['티켓값'].agg(['max', 'min', 'median']).reset_index()

# 원본 데이터프레임과 합치기
result_df = pd.merge(sale, grouped, on='공연일시', suffixes=('', '_공연일시'))
result_df = result_df[['공연코드','공연일시','max','min','median']]
result_df = result_df.drop_duplicates()
result_df
```
