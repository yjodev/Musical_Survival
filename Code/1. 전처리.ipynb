```python
import pandas as pd
import os

def process_files_in_range(file_path, file_names, start_index, last_index, target_file):
    dfs = []
    for i in range(start_index, last_index):
        file = file_names[i]
        df = pd.read_excel(os.path.join(file_path, file))
        dfs.append(df)
        print(i,' 번째 append 완')

    merged_df = pd.concat(dfs, ignore_index=True)
    merged_df.to_csv(target_file, index=False)
    print(start_index,' 부터', last_index,' 까지', target_file,'로 CSV변경 완')
```

## 3-3. 공연시설별 교통구역과의 거리(harversion distance) 계산

```python
import numbers
import math

class GeoUtil:
    """
    Geographical Utils
    """
    @staticmethod
    def degree2radius(degree):
        return degree * (math.pi/180)

    @staticmethod
    def get_harversion_distance(x1, y1, x2, y2, round_decimal_digits=5):
        """
        경위도 (x1,y1)과 (x2,y2) 점의 거리를 반환
        Harversion Formula 이용하여 2개의 경위도간 거래를 구함(단위:Km)
        """
        if x1 is None or y1 is None or x2 is None or y2 is None:
            return None
        assert isinstance(x1, numbers.Number) and -180 <= x1 and x1 <= 180
        assert isinstance(y1, numbers.Number) and  -90 <= y1 and y1 <=  90
        assert isinstance(x2, numbers.Number) and -180 <= x2 and x2 <= 180
        assert isinstance(y2, numbers.Number) and  -90 <= y2 and y2 <=  90

        R = 6371 # 지구의 반경(단위: km)
        dLon = GeoUtil.degree2radius(x2-x1)
        dLat = GeoUtil.degree2radius(y2-y1)

        a = math.sin(dLat/2) * math.sin(dLat/2) \
            + (math.cos(GeoUtil.degree2radius(y1)) \
              *math.cos(GeoUtil.degree2radius(y2)) \
              *math.sin(dLon/2) * math.sin(dLon/2))
        b = 2 * math.atan2(math.sqrt(a), math.sqrt(1-a))
        return round(R * b * 1000, round_decimal_digits) #km를 m로 표현하고자 1000 곱해줌

    @staticmethod
    def get_euclidean_distance(x1, y1, x2, y2, round_decimal_digits=5):
        """
        유클리안 Formula 이용하여 (x1,y1)과 (x2,y2) 점의 거리를 반환
        """
        if x1 is None or y1 is None or x2 is None or y2 is None:
            return None
        assert isinstance(x1, numbers.Number) and -180 <= x1 and x1 <= 180
        assert isinstance(y1, numbers.Number) and  -90 <= y1 and y1 <=  90
        assert isinstance(x2, numbers.Number) and -180 <= x2 and x2 <= 180
        assert isinstance(y2, numbers.Number) and  -90 <= y2 and y2 <=  90

        dLon = abs(x2-x1) # 경도 차이
        if dLon >= 180:   # 반대편으로 갈 수 있는 경우
            dLon -= 360   # 반대편 각을 구한다
        dLat = y2-y1      # 위도 차이
        return round(math.sqrt(pow(dLon,2)+pow(dLat,2)),round_decimal_digits)
```

```python
#행의 수가 8795, 8771, 8771, 4421이 아닌 csv 보정

import os
import pandas as pd
import shutil

def trim_csv(input_file_path, target_row_count):
    try:
        df = pd.read_csv(input_file_path)
        current_row_count = df.shape[0]

        if current_row_count != target_row_count:
            delete_from_row = target_row_count
            df = df.iloc[:delete_from_row]

            # Save the trimmed DataFrame to the new CSV file
            output_file_path = os.path.join(output_folder, os.path.basename(input_file_path))
            df.to_csv(output_file_path, index=False)

            print(f"{input_file_path} 파일을 행의 수 {target_row_count}로 수정하여 {output_file_path}에 저장했습니다.")
        else:
            # 행 수가 목표값과 일치할 경우 복사하여 저장
            output_file_path = os.path.join(output_folder, os.path.basename(input_file_path))
            shutil.copy(input_file_path, output_file_path)

            print(f"{input_file_path} 파일은 이미 목표 행의 수와 일치하여 {output_file_path}에 저장했습니다.")
    except Exception as e:
        print(f"오류 발생: {e}")

def process_csv_files(input_folder, output_folder):
    file_list = os.listdir(input_folder)
    csv_files = [file for file in file_list if file.endswith('.csv')]

    # 새로운 디렉토리 생성
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # 각 파일에 대해 행 수를 맞춰주는 코드 추가
    target_row_counts = [8795, 8771, 8771, 4421]
    for i, csv_file in enumerate(csv_files):
        target_row_count = target_row_counts[i % len(target_row_counts)]
        input_file_path = os.path.join(input_folder, csv_file)

        # 행 수를 맞추고, 수정된 파일을 새로운 디렉토리에 저장
        trim_csv(input_file_path, target_row_count)

if __name__ == "__main__":
    input_folder = "/content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/강남구_날씨_2019_2022"
    output_folder = "/content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/강남구_날씨_2019_2022_Cleaned"

    process_csv_files(input_folder, output_folder)
```

```python
import os
import pandas as pd

%cd /content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/강남구_날씨_2019_2022_Cleaned


def merge_csvs_by_group(file_list, output_folder):
    # 파일들을 두 번째 언더바까지의 정보를 기준으로 그룹화
    groups = {}
    for csv_file in file_list:
        key = '_'.join(csv_file.split('_')[:2])
        if key not in groups:
            groups[key] = []
        groups[key].append(csv_file)

    # 각 그룹별로 파일들을 병합하여 저장
    for group_key, group_files in groups.items():
        # 빈 DataFrame 생성
        merged_df = pd.DataFrame()
        # 각 파일을 순환하면서 병합
        for csv_file in group_files:
            # 첫 번째 행을 제외하고 읽어옴
            df = pd.read_csv(csv_file)
            df.columns = ["date", "time", "value"]

            # "date" 열에서 "Start"를 포함하는 행을 삭제
            df = df[~df['date'].str.contains('Start')]

            merged_df = pd.concat([merged_df, df], ignore_index=True)


        # 파일 이름에 '2019_2022'를 추가하여 저장
        output_filename = f"{output_folder}/{group_key}_2019_2022.csv"
        merged_df.to_csv(output_filename, index=False)
        print(f"{group_key} 병합 및 저장 완료: {output_filename}")
```

### 4-3. date, time format 바꾸기

```python
import os
import pandas as pd
import datetime

# Function to convert hour value to 'hh:mm' format
def convert_hour_format(time):
    if pd.notnull(time):
        try:
            hours = int(time) // 100
            minutes = int(time) % 100
            return f"{hours:02d}:{minutes:02d}"
        except ValueError:
            return 'NA'  # Return 'NA' if hour value is not a valid integer
    return 'NA'  # Return 'NA' for NaN values

# Function to perform the column renaming based on filename
def rename_columns(df, filename):
    df.rename(columns={df.columns[0]: 'date', 'time': 'time_temp'}, inplace=True)
    df['time'] = df['time_temp'].apply(convert_hour_format)
    df.drop(columns='time_temp', inplace=True)
    return df

# Function to modify the 'date' column with dates from 2019-07-01 to 2022-12-31
def modify_date_column(df):
    start_date = datetime.date(2019, 7, 1)
    end_date = datetime.date(2022, 12, 31)
    dates = [start_date + datetime.timedelta(days=i) for i in range((end_date - start_date).days + 1)]
    df['date'] = sorted(dates * 24)  # Sort the dates and repeat 24 times for each date
    return df

# Directory where the files are located
directory = '/content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/merged_마포구_날씨_2019_2022'
out_directory = '/content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/format_마포구_날씨_2019_2022'
os.makedirs(out_directory, exist_ok=True)

# Get the list of files in the directory
file_list = os.listdir(directory)

# Iterate through each file
for file in file_list:
    if file.endswith('.csv'):
        # Read the file into a DataFrame
        df = pd.read_csv(os.path.join(directory, file))

        # Get the filename without extension
        filename_without_extension = os.path.splitext(file)[0]

        # Perform column renaming based on filename
        df = rename_columns(df, filename_without_extension)

        # Reorder columns to have 'hour' as the second column
        if 'time' in df.columns:
            cols = list(df.columns)
            cols.insert(1, cols.pop(cols.index('time')))
            df = df[cols]

        # Modify the 'date' column with dates from 2019-07-01 to 2022-12-31
        df = modify_date_column(df)

        # Save the updated DataFrame back to the file
        df.to_csv(os.path.join(out_directory, file), index=False)
```

```python
import os
import pandas as pd

# 데이터가 있는 경로를 지정합니다.
data_path = '/content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/format_마포구_날씨_2019_2022'  # 데이터가 있는 경로로 변경해주세요.

# 데이터를 저장할 결과 디렉토리를 생성합니다.
result_path = '/content/drive/MyDrive/KOPIS DATA/Data/Outsource/Weather/final_마포구_날씨_2019_2022'  # 결과를 저장할 경로로 변경해주세요.
os.makedirs(result_path, exist_ok=True)

# 첫 번째 언더바 전까지의 문자가 일치하는 CSV 파일들을 그룹화합니다.
grouped_files = {}
files = os.listdir(data_path)
files.sort()  # Sort filenames
for csv_file in files:
    file_name = csv_file.split('.')[0]
    group_name = '_'.join(file_name.split('_')[:1])
    if group_name not in grouped_files:
        grouped_files[group_name] = []
    grouped_files[group_name].append(csv_file)

# 그룹화된 파일들을 읽어들여서 결과 데이터프레임에 세 번째 열을 추가합니다.
result_df = pd.DataFrame()
for group_name, files in grouped_files.items():
    group_dfs = []
    first_file_read = False
    for idx, csv_file in enumerate(files):
        csv_path = os.path.join(data_path, csv_file)
        df = pd.read_csv(csv_path)
        column_name = csv_file.split('_')[1]  # 첫 번째 언더바와 두 번째 언더바 사이의 문자를 가져옵니다.
        if not first_file_read:
            # 첫 번째 파일의 경우 세 번째 열 이름을 변경합니다.
            df.columns.values[2] = column_name
            group_dfs.append(df)
            first_file_read = True
        else:
            group_dfs.append(df.iloc[:, 2].rename(column_name))
    merged_df = pd.concat(group_dfs, axis=1)
    result_df = pd.concat([result_df, merged_df], axis=0)

    # 새로운 파일 이름을 지정하여 결과를 저장합니다.
    new_file_name = f"{group_name}_강수기온습도_2019_2022.csv"
    result_file_path = os.path.join(result_path, new_file_name)
    merged_df.to_csv(result_file_path, index=False)

print("작업이 완료되었습니다. 결과는", result_path, "에 저장되었습니다.")
```

```python
import re

def convert_to_minutes(time_str):
    if pd.notna(time_str):
        hours = re.findall(r'\d+시간', time_str)
        minutes = re.findall(r'\d+분', time_str)

        total_minutes = 0
        if hours:
            total_minutes += int(hours[0].replace('시간', '')) * 60
        if minutes:
            total_minutes += int(minutes[0].replace('분', ''))

        return total_minutes
    else:
        return None
```
